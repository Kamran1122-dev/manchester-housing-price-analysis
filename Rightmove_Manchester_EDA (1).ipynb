This project explores housing price patterns in Manchester using a synthetic but realistic dataset.

```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("manchester_housing_data.csv")
df.head()
```

### Summary Statistics

```
df.describe()
```

### Price Distribution

```

plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=30)
plt.xlabel('Price (Â£)')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')
plt.show()
```

### Correlation Heatmap

```

plt.figure(figsize=(7,5))
sns.heatmap(df.corr(), annot=True)
plt.title('Feature Correlation Heatmap')
plt.show()
```

### Average Price by Postcode

```

postcode_price = df.groupby('postcode')['price'].mean().sort_values()
postcode_price
```
